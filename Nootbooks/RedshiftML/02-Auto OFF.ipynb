## AUTO:OFF syntax

```redshift
CREATE MODEL model_name
    FROM { table_name | (select_statement ) }
    TARGET column_name
    FUNCTION function_name
    IAM_ROLE { default }
    AUTO OFF
    MODEL_TYPE XGBOOST
    OBJECTIVE { 'reg:squarederror' | 'reg:squaredlogerror' | 'reg:logistic' |
                'reg:pseudohubererror' | 'reg:tweedie' | 'binary:logistic' | 'binary:hinge' |
                'multi:softmax' | 'rank:pairwise' | 'rank:ndcg' }
    HYPERPARAMETERS DEFAULT EXCEPT (
        NUM_ROUND '10',
        ETA '0.2',
        NUM_CLASS '10',
        (, ...)
    )
    PREPROCESSORS 'none'
    SETTINGS (
      S3_BUCKET 'bucket', |
      S3_GARBAGE_COLLECT { ON | OFF }, |
      KMS_KEY_ID 'kms_key_id', |
      MAX_CELLS integer, |
      MAX_RUNTIME integer (, ...)
    )
```

### Inspect the data

```redshift
select * from <<YOUR_SCHEMA_NAME>>.<<YOUR_TABLE_NAME>> limit 10;
```

### AUTO:OFF Example

```redshift
CREATE MODEL <<YOUR_MODLE_NAME>>
FROM <<YOUR_SCHEMA_NAME>>.<<YOUR_TABLE_NAME>>
TARGET <<YOUR_TARGET_COLUMN>>
FUNCTION <<YOUR_FUNCTION_NAME>>
IAM_ROLE ''<<YOUR_IAM_ROLE_ARN>>'
AUTO OFF
MODEL_TYPE xgboost
OBJECTIVE 'binary:logistic'
PREPROCESSORS 'none'
HYPERPARAMETERS DEFAULT EXCEPT(
    ALPHA '3.948832276888859e-05',
    ETA '0.09007405390404363',
    COLSAMPLE_BYTREE '1.0',
    EVAL_METRIC 'f1',
    GAMMA '0.3819349917976025',
    LAMBDA '0.21294732751634948',
    MAX_DEPTH '8',
    MIN_CHILD_WEIGHT '0.009041488538152996',
    NUM_ROUND '20',
    SCALE_POS_WEIGHT '7.876724137931035', 
    SUBSAMPLE '1.0')
SETTINGS(
  S3_BUCKET '<<YOUR_BUCKET_NAME>>',
  MAX_RUNTIME 900
);
```

### Describe Model:

```redshift
SHOW MODEL <<YOUR_SCHEMA_NAME>>.<<YOUR_MODEL_NAME>>
```

### Perform predictions

```redshift
SELECT predicted_result, original_result,COUNT(*) 
  FROM( 
      SELECT <<YOUR_SCHEMA_NAME>>.<<YOUR_FUNCTION_NAME>>(age,duration,campaign,pdays,previous,emp_var_rate,cons_price_idx,cons_conf_idx,euribor3m,nr_employed,job_index,marital_index,education_index,default_index,housing_index,loan_index,contact_index,month_index,day_of_week_index,poutcome_index) predicted_result, y_index as original_result 
        FROM <<YOUR_SCHEMA_NAME>>.<<YOUR_TABLE_NAME>> )
  GROUP BY predicted_result, original_result;
```
