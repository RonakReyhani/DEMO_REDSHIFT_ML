## Bring Your Amazon SageMaker model

![General Overview](../../assets/remote-inference.png)

### Remote Inference syntax

```redshift
CREATE MODEL model_name 
    FUNCTION function_name ( data_type [, ...] )
    RETURNS data_type
    SAGEMAKER 'endpoint_name'[:'model_name']
    IAM_ROLE { default | 'arn:aws:iam::<account-id>:role/<role-name>' }
```

### Remote inference Example

```redshift
CREATE MODEL '<<YOUR_SCHEMA_NAME>>.<<YOUR_MODEL_NAME>>'
FUNCTION <<YOUR_FUNCTION_NAME>> (int,int,int,int,int,float,float,float,float,float,int,int,int,int,int,int,int,int,int,int)
RETURNS int
SAGEMAKER '<<YOUR_SAGEMAKER_ENDPOINT_NAME>> '
IAM_ROLE '<<YOUR_IAM_ROLE_ARN>>'
```

### Describe Model

```redshift
SHOW MODEL <<YOUR_SCHEMA_NAME>>.<<YOUR_MODEL_NAME>>
```

### Perform predictions

```redshift
SELECT prediction, original, COUNT(*)
  FROM (SELECT <<YOUR_SCHEMA_NAME>>.<<YOUR_FUNCTION_NAME>>(
                   age,duration,campaign,pdays,previous,emp_var_rate,cons_price_idx,cons_conf_idx,euribor3m,nr_employed,job_index,marital_index,education_index,default_index,housing_index,loan_index,contact_index,month_index,day_of_week_index,poutcome_index), y_index as original
        FROM demo_ml.xgboost)
GROUP BY prediction, original;
```
