# How to use Redshift for Machine Learning

![General Overview](../../assets/auto.png)

### Inspect the data

```redshift
select * from <<YOUR_SCHEMA_NAME>>.<<YOUR_TABLE_NAME>> limit 10;
```

### Auto ON Syntax

```redshift
CREATE MODEL <<YOUR_MODLE_NAME>>
    FROM { table_name | ( select_query ) }
    TARGET column_name
    FUNCTION prediction_function_name
    IAM_ROLE { default }
    SETTINGS (
      S3_BUCKET 'bucket',
      [ MAX_CELLS integer ]
    )
```

### Auto ON Examples

#### Train model with all columns

```redshift
CREATE MODEL <<YOUR_SCHEMA_NAME>>.<<YOUR_MODEL_NAME>>
FROM <<YOUR_SCHEMA_NAME>>.<<YOUR_TABLE_NAME>>
TARGET <<YOUR_TARGET_COLUMN>>
FUNCTION <<YOUR_FUNCTION_NAME>>
IAM_ROLE '<<YOUR_IAM_ROLE_ARN>>'
SETTINGS (
  S3_BUCKET '<<YOUR_BUCKET_NAME>>'
);
```

#### Train model with subset of columns

```redshift
CREATE MODEL  <<YOUR_SCHEMA_NAME>>.<<YOUR_MODEL_NAME>>
FROM (
      SELECT age, job, marital, education, housing, contact, month, day_of_week, y FROM <<YOUR_SCHEMA_NAME>>.<<YOUR_TABLE_NAME>>
)
TARGET y
FUNCTION <<YOUR_FUNCTION_NAME>>
IAM_ROLE '<<YOUR_IAM_ROLE_ARN>>'
SETTINGS (
  S3_BUCKET '<<YOUR_BUCKET_NAME>>'
);
```

### Describe Model:

```redshift
SHOW MODEL <<YOUR_SCHEMA_NAME>>.<<YOUR_MODEL_NAME>>
```

### Model explainability

```redshift
select EXPLAIN_MODEL ('<<YOUR_SCHEMA_NAME>>.<<YOUR_MODEL_NAME>>')
```

### Perform predictions

```redshift
SELECT prediction, original, COUNT(*)
  FROM (SELECT <<YOUR_SCHEMA_NAME>>.<<YOUR_FUNCTION_NAME>>(
                   age, job, marital, education, credit_default, housing,
                   loan, contact, month, day_of_week, duration, campaign,
                   pdays, previous, poutcome, emp_var_rate, cons_price_idx,
                   cons_conf_idx, euribor3m, nr_employed), y
          FROM <<YOUR_SCHEMA_NAME>>.<<YOUR_TABLE_NAME>>)
  GROUP BY prediction, original;
```

```redshift
DROP MODEL '<<YOUR_SCHEMA_NAME>>.<<YOUR_MODEL_NAME>>';
```
